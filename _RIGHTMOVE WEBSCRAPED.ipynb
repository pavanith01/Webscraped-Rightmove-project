```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

```

```python
## create this to get all the data into a dataframe with pandas
data = []
```

```python
total_pages = 5  # Number of pages to scrape from website 
```

```python
 ## url varies between tasks 
    
    url = "https://www.rightmove.co.uk/property-to-rent/find.html?locationIdentifier=REGION%5E93947&maxBedrooms=2&minBedrooms=2&maxPrice=2000&minPrice=1500&propertyTypes=&includeLetAgreed=false&mustHave=&dontShow=&furnishTypes=&keywords="
##this part of the code allows webscraping for other pages rather than just 1 page
## response is supposed to be 200 so make sure to check because 200 will allow information to eb craped of the website 
for page in range(1, total_pages + 1):
    response = requests.get(url + f"&index={page}")
    soup = BeautifulSoup(response.content, 'html.parser')

    ## this was done just to get an idea of which class and div/span to use to get the exact price or address values only 
    property_listings = soup.find_all('div', class_='propertyCard')   
    
    ## sometimes not everything can be scraped of the website and it can vary.
    for listing in property_listings:
        price = listing.find('span', class_='propertyCard-priceValue').text.strip()
        address = listing.find('address').text.strip()
        
        ## scraped description to add rest of information as there could be a garden loft etc.
        additional_info = ""
        info_elements = listing.find_all('div', class_='propertyCard-description')
        for info_element in info_elements:
            additional_info += info_element.text.strip() + " "
        
        ## data need to be appended to get a new dataframe
        data.append([price, address, additional_info])
        
        
```

```python

# Convert the data into a Pandas DataFrame
df = pd.DataFrame(data, columns=['Price', 'Address', 'Additional Information'])

```
